```python
from ultralytics import YOLO
import os
import shutil
from google.colab import files

# ✅ Configuration
DATASET_PATH = "/content/merged_dataset/dataset.yaml"  # Ensure this exists!
PROJECT_DIR = "/content/drive/MyDrive/yolo_training"
RUN_NAME = "run1"
MODEL_SAVE_DIR = f"{PROJECT_DIR}/{RUN_NAME}/weights"
LOCAL_DOWNLOAD_DIR = "/content"  # Local directory for downloads

# ✅ Training parameters
EPOCHS = 100
BATCH_SIZE = 8
IMG_SIZE = 640
DEVICE = "cuda"
WORKERS = 2
SAVE_PERIOD = 5  # Save model every 5 epochs
PATIENCE = 15  # 🔹 Stops training if no improvement for 10 consecutive epochs

# 1️⃣ Find the latest checkpoint
CHECKPOINT_PATH = os.path.join(MODEL_SAVE_DIR, "last.pt")  # Default checkpoint

if os.path.exists(CHECKPOINT_PATH):
    print(f"🔄 Resuming training from last checkpoint: {CHECKPOINT_PATH}")
    model = YOLO(CHECKPOINT_PATH)  # Resume from last saved checkpoint
    resume_training = True
else:
    print("🚀 Starting fresh training from 'yolov8s.pt'")
    model = YOLO("yolov8s.pt")  # Start from scratch
    resume_training = False

# 2️⃣ Resume training with early stopping
try:
    model.train(
        data=DATASET_PATH,
        epochs=EPOCHS,
        batch=BATCH_SIZE,
        imgsz=IMG_SIZE,
        device=DEVICE,
        workers=WORKERS,
        amp=True,  # Mixed precision for faster training
        project=PROJECT_DIR,
        name=RUN_NAME,
        save_period=SAVE_PERIOD,  # Saves model every 5 epochs
        patience=PATIENCE,  # ⏳ Stops training if no improvement for 10 epochs
        resume=resume_training  # 🔄 Automatically resumes if checkpoint exists
    )
    print("✅ Training resumed successfully!")
except Exception as e:
    print(f"⚠ Training failed: {e}")
    raise

# 3️⃣ Evaluate the trained model on the test set
try:
    metrics = model.val(
        split="test",
        project=PROJECT_DIR,
        name=RUN_NAME
    )

    # ✅ Extract evaluation metrics
    map_50_95 = metrics.box.map
    map_50 = metrics.box.map50
    map_75 = metrics.box.map75
    precision = metrics.box.p
    recall = metrics.box.r

    print("\n📊 Model Evaluation Metrics:")
    print(f"🔹 mAP@50-95: {map_50_95:.4f}")
    print(f"🔹 mAP@50: {map_50:.4f}")
    print(f"🔹 mAP@75: {map_75:.4f}")
    print(f"🔹 Precision: {precision:.4f}")
    print(f"🔹 Recall: {recall:.4f}")
except Exception as e:
    print(f"⚠ Evaluation failed: {e}")
    raise

# 4️⃣ Move model to /content before downloading
MODEL_FILES = ["best.pt", "last.pt"]

for filename in MODEL_FILES:
    src_path = os.path.join(MODEL_SAVE_DIR, filename)
    dest_path = os.path.join(LOCAL_DOWNLOAD_DIR, filename)

    if os.path.exists(src_path):
        try:
            shutil.copy(src_path, dest_path)
            files.download(dest_path)  # ✅ Download model to local machine
            print(f"✅ {filename} downloaded successfully.")
        except Exception as e:
            print(f"⚠ Error downloading {filename}: {e}")
    else:
        print(f"⚠ File not found: {src_path}")
```

```python
from ultralytics import YOLO

# ✅ Load the trained model
MODEL_PATH = "/content/drive/MyDrive/yolo_training/run1/weights/best.pt"  # Update path if needed
model = YOLO(MODEL_PATH)

# ✅ Run validation on test set
metrics = model.val(split="test", project="/content/drive/MyDrive/yolo_training", name="run1")

# ✅ Extract evaluation metrics
map_50_95 = metrics.box.map          # mAP@50-95
map_50 = metrics.box.map50           # mAP@50
map_75 = metrics.box.map75           # mAP@75
precision = metrics.box.mp         # Mean Precision
recall = metrics.box.mr           # Mean Recall
f1_score = metrics.box.f1            # F1-score per class

# ✅ Print results
print("\n📊 Model Evaluation Metrics:")
print(f"🔹 mAP@50-95: {map_50_95:.4f}")
print(f"🔹 mAP@50: {map_50:.4f}")
print(f"🔹 mAP@75: {map_75:.4f}")
print(f"🔹 Precision: {precision:.4f}")
print(f"🔹 Recall: {recall:.4f}")

# 🔹 Print class-wise results
print("\n📌 Class-wise Performance:")
for i, class_name in enumerate(["axe", "knife", "scissor", "screwdriver"]):
    class_precision, class_recall, class_ap50, class_map50_95 = metrics.box.class_result(i)
    print(f"  🏷 {class_name}:")
    print(f"    ✅ Precision: {class_precision:.4f}")
    print(f"    ✅ Recall: {class_recall:.4f}")
    print(f"    ✅ AP@50: {class_ap50:.4f}")
    print(f"    ✅ mAP@50-95: {class_map50_95:.4f}")

print("\n✅ Model testing completed successfully!")
```
